```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train,y_train)
lr_clf.score(X_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

cross_val_score(LinearRegression(), X, y, cv=cv)
```

```python
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(X,y):
    algos = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
                'normalize': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(X,y)
```

```python
def predict_car_price(brand, interior, exterior, age, miles, accidents, owners):
    # Strip spaces from the inputs to ensure consistency
    brand = brand.strip()
    interior = interior.strip()
    exterior = exterior.strip()
    
    # Construct column names with the correct spacing as per your DataFrame
    brand_col = f'brand_{brand}'  # Assuming no extra spaces for brand columns
    interior_col = f'interior_ {interior} '  # Space after "interior_" and at the end, as per your provided columns
    exterior_col = f'exterior_{exterior} '  # No space after "exterior_" but a space at the end

    # Check if the constructed feature names are in the DataFrame columns
    missing_columns = [col for col in [brand_col, interior_col, exterior_col] if col not in X.columns]
    if missing_columns:
        raise ValueError(f'The feature column(s) {", ".join(missing_columns)} is/are not found in the DataFrame.')

    # Find the column indices for the brand, interior, and exterior features
    brand_index = np.where(X.columns == brand_col)[0][0]
    interior_index = np.where(X.columns == interior_col)[0][0]
    exterior_index = np.where(X.columns == exterior_col)[0][0]

    # Create an input array with zeros for all features
    x = np.zeros(len(X.columns))

    # Assuming 'age', 'mileage', 'accidents', and 'owners' columns exist without extra spaces
    age_index = np.where(X.columns == 'age')[0][0]
    mileage_index = np.where(X.columns == 'miles')[0][0]
    accidents_index = np.where(X.columns == 'accidents')[0][0]
    owners_index = np.where(X.columns == 'owners')[0][0]

    # Set values for age, mileage, accidents, and owners
    x[age_index] = age
    x[mileage_index] = miles
    x[accidents_index] = accidents
    x[owners_index] = owners

    # Set the values for brand, interior, and exterior to 1 to indicate their presence
    x[brand_index] = 1
    x[interior_index] = 1
    x[exterior_index] = 1

    # Predict the price using the trained model and return it
    return lr_clf.predict([x])[0]
```

```python
predict_car_price('Toyota', 'Black', exterior='Red', age=3, miles=50000, accidents=0, owners=1)

```

```python
predict_car_price('BMW', 'Black', exterior='White', age=3, miles=50000, accidents=0, owners=1)
```
